<h2>Pitch Sequences of Frequency Ratios</h2>

```python
# Make an equal temparement sequence of frequency ratios
rats_eq = 2**(arange(13.0)/12.0)
print rats_eq
plot_abs(rats_eq)
```

```python
# Lookup equal temparement ratios at 'maj' scale positions
print rats_eq[abs_scale['maj']]
# Lookup equal temparement ratios at 'mel_up' scale positions
print rats_eq[abs_scale['mel_up']]
```

```python
plot_abs(rats_eq)
title('Sequence of chromatic frequency ratios')
```

```python
plot_abs(rats_eq[abs_scale['mel_up']])
title('Sequence of Ascending Melodic Minor Frequency Ratios')
```

<h2>Complex Valued Coding of Relative Pitch and Rhythmic Phase</h2>

```python
d = 8 # greatest common divisor for segment (bar)
rel_phase = (cumsum(ones(d)) - 1) / d
print rel_phase
plot_angle(exp(1j*2*pi*rel_phase))
```

```python
# Lookup frequency ratios of a shape in the ascending melodic minor mode
shape = [0,1,2,1,2,3,2,3]
mel_rats = rats_eq[abs_scale['mel_up'][shape]]
print mel_rats
plot_abs(mel_rats)
```

```python
complex_seq = seq_to_complex(mel_rats,rel_phase) # convert magnitudes (ratios) and phase angles to complex form
plot_seq(complex_seq)
for z in complex_seq-seq_to_complex(*complex_to_seq(complex_seq)): # identity test
    print abs(z)
```

```python
seq2 = phase_scale(complex_seq, 0.5)
seq2 = phase_offset(seq2, 0.0625) # rhythmic augmentation by 2, phase_offset 0.0625
print relative_phase(complex_seq)
print relative_phase(seq2)

subplot(221)
plot_abs(complex_seq)
subplot(222)
plot_angle(complex_seq) 
subplot(223)
plot_abs(seq2)
subplot(224)
plot_angle(seq2) 
```

```python
pitch_transpose = lambda x, trans: x*trans
pitch_transpose (complex_seq, 3./2.) # up a perfect 5th
plot_seq(pitch_transpose (complex_seq, 3./2.))
```

```python
print relative_phase(complex_seq)
plot_seq(complex_seq)
```

```python
shape1 = seq_to_complex([1.,3/2.,4/3.,1.],array([0,0.25,0.5,0.75])) # a shape is a list of target values
struct1 = kron(shape1,[1,0,0,0,0,0,0,0]) # a structure (struct) is a sparse shape
melody1 = convolve(struct1, seq_to_complex([1.0,3/2.,4/3.,1.0],array([0.,0.0625,0.125,0.1875])/2.)) # a sequence is a non-sparse shape
plot_seq(melody1)
```
